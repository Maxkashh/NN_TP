# Процессинг аудио файлов

Теперь нам нужны только те части из всего аудиофайла, которые содержат дыхательные циклы.
Мы можем сделать это, используя параметры времени начала и окончания для этих циклов в нашем фрейме данных.



```python
def getPureSample(raw_data,start,end,sr=22050):
    max_ind = len(raw_data) 
    start_ind = min(int(start * sr), max_ind)
    end_ind = min(int(end * sr), max_ind)
    return raw_data[start_ind: end_ind]
```

мы умножили start на частоту дискретизации, потому что start — это время, а raw_data — это выборка массива в соотв. к частоте дискретизации

Теперь мы также хотим, чтобы наши входные изображения для нашего cnn были одинакового размера, поскольку аудиофайлы должны быть одинаковой длины т.е. (начало-конец) должны быть одинаковыми

```python
import soundfile as sf
import librosa as lb
i,c=0,0
for index,row in data.iterrows():
    maxLen=6
    start=row['start']
    end=row['end']
    filename=row['filename']
    
    #If len > maxLen , меняем на  maxLen
    if end-start>maxLen:
        end=start+maxLen
    
    audio_file_loc=path + filename + '.wav'
    
    if index > 0:
        #проверьте, выходит ли больше циклов для одного и того же пациента, если да, то добавьте i, чтобы изменить имя файла
        if data.iloc[index-1]['filename']==filename:
            i+=1
        else:
            i=0
    filename= filename + '_' + str(i) + '.wav'
    
    save_path='processed_audio_files/' + filename
    c+=1
    
    audioArr,sampleRate=lb.load(audio_file_loc)
    pureSample=getPureSample(audioArr,start,end,sampleRate)
    
    #pad audio if pureSample len < max_len
    reqLen=6*sampleRate
    padded_data = lb.util.pad_center(pureSample, reqLen)
    
    sf.write(file=save_path,data=padded_data,samplerate=sampleRate)
print('Всего обработано файлов: ',c)
```

```python
def get_features(path):
    sound_arr,sample_rate=lb.load(path)
    mfcc=lb.feature.mfcc(y=sound_arr,sr=sample_rate)
    cstft=lb.feature.chroma_stft(y=sound_arr,sr=sample_rate)
    melspec=lb.feature.melspectrogram(y=sound_arr,sr=sample_rate)

    return mfcc,cstft,melspec
```
